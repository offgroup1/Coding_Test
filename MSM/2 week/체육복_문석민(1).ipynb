```python
n = 5
lost = [1,3, 2,4]
reserve = [5, 4, 2]

## 우선적으로 lost, reserve list를 중복값을 기준으로 정렬시킨다.
## 두번 동일한 프로세스를 진행해야하기때문에 함수를 정의하여 사용한다.

def sorted_lst(lst1, lst2):
    f_lst = []
    s_lst = []

    for elem in lst2:
        if elem in lst1:
            f_lst.append(elem)
        else:
            s_lst.append(elem)
    f_lst.sort()
    s_lst.sort()

    return f_lst + s_lst
reserve = sorted_lst(lost, reserve)
lost = sorted_lst(reserve, lost)


for elem in reserve:
    if elem in lost:
        lost.remove(elem)
    elif elem - 1 in lost:
        lost.remove(elem - 1)
    elif elem + 1 in lost:
        lost.remove(elem + 1)

print(n - len(lost))
```
