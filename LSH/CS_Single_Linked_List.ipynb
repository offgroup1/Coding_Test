```python
class Node:
  def__init__(self,data,next=None):
    self.data = data
    self.next = next
  
  def delete(self, data):
    if self.head == '':
      print ("해당 값을 가진 노드가 없습니다.")
      return
    #1    
    if self.head.data == data: #데이터가 있을 때 까지
        temp = self.head #지금 현재 위치한 노드
        self.head = self.head.next #원래 헤드 다음 노드로 헤드를 바꿈
        del temp #전 헤드 지우기
    #2&3
    else:
        node = self.head
        while node.next:
            if node.next.data == data: #다음 데이터가 삭제할 데이터인가?
                temp = node.next #맞으면 
                node.next = node.next.next #그 다음 노드로 넘겨주기
                del temp #삭제
                return
            else:
                node = node.next #찾는 노드가 아니면 이동
```

단점

- 연결을 위한 별도의 데이터 공간이 필요하므로, 저장공간 효율이 높지 않음
