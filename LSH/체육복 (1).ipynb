```python
#1.처음엔 모두가 한개씩 가지고 있다고 가정해야 하니깐 1 로 저장된 n+2개의 원소를 가진 리스트 만들기
#2.for 문을 돌면서 빼고 더해주기
#3.큰수쪽으로 살필 지 작은 수 쪽으로 살필지 정하기
#4.조건에 해당하면 둘다 1로 바꿔주기
#5.리스트 안에 0 보다 큰 수 를 가진 수  

def solution(n, lost, reserve):
    one = [1]*(n+2) #1
    for i in reserve: #2
        one[i] += 1
    for i in lost:
        one[i] -= 1
    for i in range(1,n+1):
        if one[i-1] == 0 and one[i] == 2: #3
            one[i-1:i+1] = [1,1] #4
        elif one[i] == 2 and one[i+1] == 0:
            one[i:i+2] = [1,1]
    
    return len([x for x in one[1:-1] if x > 0])#5
```

```python
def solution(n, lost, reserve):
    s = set(lost) & set(reserve) #교집합
    l = set(lost) - s #차집합: 0개인 친구들만
    r = set(reserve) - s #2개인 친구들만
    for x in sorted(r): 
        if x-1 in l:
            l.remove(x-1)
        elif x+1 in l:
            l.remove(x+1)
    return n - len(l)
```
