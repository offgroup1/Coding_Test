```python
#이번 문제는 이진탐색이기에 대상 설정이 가장 중요했습니다
#저는 결국 걸리는  시간을 초점으로 이진 탐색을 했습니다.
1#최소시간/ 최대시간: 심사대에 모든 사람이 들어감

def solution(n,times):
    answer = max(times) * n
    left = 1 #최소시간
    right = answer #심사대에 모든 사람이 다 들어가서 걸리게 되는  시간
    while left < right: #최소시간을 구하는 것이므로 left가 right를 넘어가기 전까지 합니다.
        mid = (left+right) //2 #중간값
        # 중간값 시간 내에 n명을 모두 심사할 수 있는가
        judge = 0
        for t in times:
            judge += mid //t
        #더 짧은 시간이 가능한가
        if judge >n:
            right = mid
            if mid < answer:
                answer = mid
        #불가능하다면 최소 시간을 늘려서 다시 시작
        else:
            left = mid + 1
    return answer

 
```
